1. Write a function that lists all the possible outcomes with five dice.

This algorithm calculates all possible outcomes with respect to the order first and then takes the subset to find the outcomes (another option would be to directly calculate just the relevant outcomes :D)

The subset is taken by first sorting all the elements in the list of Outcomes. tuple() makes them immutable. Then i take the set because the set cannot contain multiple times the same element.


```python
# function that lists all possible outcomes for 5 dice
# order doesn't matter: n=6, k=5, number of possible outcomes = (5+6-1)!/(5!(10-5)!) = 252

def getPossibleOutcomes5Dice ():
    numOfDiceRolls = 5
    possibleOutcomes = [[]] # track all the possible outcomes

    # For each time we throw a dice we add all the possible outcomes (1-6) to the already calculated (intermediate) outcome
    # this is an iterative approach (another approach would be a recursive function)
    for throw in range(numOfDiceRolls):
        # store the intermediate outcomes in a temporary copy
        possibleOutcomesCopy = possibleOutcomes.copy()
        possibleOutcomes = [] 
        for intermediateOutcome in possibleOutcomesCopy:
            for outcome in [1,2,3,4,5,6]:
                # all possible outcomes of one dice roll are appended to the intermediate Outcome 
                possibleOutcomes.append(intermediateOutcome + [outcome])
    # if the order of dice would matter we would have 7776 possible outcomes
    print("Number of possible outcomes (with order):", len(possibleOutcomes))
        
    # return a subset of the total number of outcomes excluding the permutations of each combination because order doesn't matter
    print("Number of possible outcomes (without order):",len(set(tuple(sorted(subset)) for subset in possibleOutcomes)))
    return(set(tuple(sorted(subset)) for subset in possibleOutcomes))
            
    
getPossibleOutcomes5Dice()

    
```

Write functions that return, given five dice, ​True​ or ​False​ if we have Straight, Full,Poker or Generala

```python
# functions that return a boolean according to if we have a straight, full, poker or generala

num = [5,5,5,5,5]


# first create function that checks if numbers in the list are in between 1 and 6
def isDiceRoll (dices):
    for i in dices:
        if i < 1 or i > 6 :
            return False
    return True



# in each function i first check if the list of "dices" is actually a dice roll, if not give error message

# isStraight() returns true if we have combination of five consecutive numbers (the transition from 6 to 1 counting)
def isStraight (dices):
    if not isDiceRoll (dices) :
        print("Is not a dice roll") # normally you could return an Error-message here
    for i in dices:
        # if one element occurs more than one time it cannot be a straight
        if dices.count(i) > 1 :
            return False 
     # if all elements occur just one time it is a straight
    return True

# isFull() returns true if we have a set of three combined with a set of two
def isFull (dices):
    if not isDiceRoll (dices) :
        print("Is not a dice roll")
    for i in dices:
        # if one element just occurs 1 time or over 3 times it cannot be a full
        if dices.count(i) < 2 or dices.count(i) > 3 :
            return False
    return True
   
# isPoker() returns true if we have four dice with the same number        
def isPoker (dices):
    if not isDiceRoll (dices) :
        print("Is not a dice roll")
    for i in dices:
        # if one of the elements occurs 4 times it has to be a Poker
        if dices.count(i) == 4 :
            return True
    return False

# isGenerala() returns true if we have all five dice with the same number
def isGenerala (dices) :
    if not isDiceRoll (dices) :
        print("Is not a dice roll")
    for i in dices:
        # if one element occurs 5 times it has to be a Generala
        if dices.count(i) == 5 :
            return True
    return False
        
isGenerala(num)
```

Compute the probabilities that you already computed by hand by counting the outcomes describing each roll result in the list of all possible hands served.

```python
# calculate the number of outcomes that fullfill a Straight, Full,Poker and Generala (served)
# The probabilities are simply the number of the different hands divided by the number of possible outcomes

# probOfHandsServ(handType) takes a function that calculates if a hand is a certain type (straight,poker) that returns 
# a boolean, so for example isStraight, isPoker,....

def probOfHandServ(handType) :
    counter = 0
    possibleOutcomes = getPossibleOutcomes(5)
    for i in possibleOutcomes :
        if handType(i) :
            counter += 1
    return counter/len(possibleOutcomes)

print("Prob. of possible Straights served: ",probOfHandServ(isStraight))
print("Prob. of possible Fulls served: ",probOfHandServ(isFull))
print("Prob. of possible Pokers served: ",probOfHandServ(isPoker))
print("Prob. of possible Generalas served: ",probOfHandServ(isGenerala))
```

5. Compute the probabilities that you already computed by hand by counting the outcomes describing each roll result in the list of all possible hands.

In each game of Generala we have three rounds to play and throw the dice.
The possible hands of outcomes stay the same in each round, so even though you can keep a few dice and just
roll a subset of dice the probability of receiving a certain roll result stays the same.
So the overall probabilities for each outcome (straight, poker, ..) can just be multiplied by 3.

```python
# probOfHands(handType) takes a function that calculates if a hand is a certain type (straight,poker) that returns 
# a boolean
def probOfHands (handType) :
    return probOfHandServ(handType) * 3


print("Prob. of possible Straights: ",probOfHands(isStraight))
print("Prob. of possible Fulls: ",probOfHands(isFull))
print("Prob. of possible Pokers: ",probOfHands(isPoker))
print("Prob. of possible Generalas: ",probOfHands(isGenerala))
```
